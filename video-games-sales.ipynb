```python
# Read data
data = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')

# Set style
sns.set_style("whitegrid")

# Plotting
plt.figure(figsize=(10, 8))
scatter_plot = sns.scatterplot(x='NA_Sales', y='EU_Sales', data=data, hue='Platform', palette='Set2', alpha=0.8, s=100)
plt.title('NA Sales vs EU Sales', fontsize=16, fontweight='bold')
plt.xlabel('NA Sales (in millions)', fontsize=12)
plt.ylabel('EU Sales (in millions)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Platform', loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Adjust legend
legend = scatter_plot.legend()
legend.set_title('Platform', prop={'size': 10, 'weight': 'bold'})

plt.show()
```

```python
pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')

# Group by Year and sum Global Sales
sales_over_years = data.groupby('Year')['Global_Sales'].sum()

# Plotting
plt.figure(figsize=(10, 6))
sales_over_years.plot(kind='line', marker='o', color='green')
plt.title('Global Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=data)
plt.title('Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.hist(data['Global_Sales'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
